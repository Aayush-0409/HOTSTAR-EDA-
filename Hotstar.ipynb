```python

#mean, median - replace numberic values (prefer median if data has outliers)
#mode, ffill, bfill - replace categorical data

#fillna() - is used to fill null values

df['running_time'] = df["running_time"].fillna(df['running_time'].median())
df
```

```python
#check for duplicates

df.duplicated().sum()
```

```python
df = df.drop_duplicates() #to remove the duplicates
```

```python
#to visualize the count of tv shows and movies
sns.countplot(x =df["type"], data = df , color = "aqua")
plt.title("Count of TV shows and Movies") #gives title to our chart
plt.show() #to display the chart
```

```python
gb = df.groupby("type").agg({"hotstar_id":"count"}) #similar to pivots in excel
gb
```

```python
plt.figure(figsize = (10,4))
sns.countplot(x = df["genre"],data = df, color="blue")
plt.title("Count of genre")
plt.xticks(rotation=90)
plt.show()
```

```python
gb = df.groupby("age_rating").agg({"hotstar_id":"count"})
gb
```

```python
plt.pie(gb["hotstar_id"], labels = gb.index, autopct = "%.2f%%") # type: ignore
plt.show()
```

```python
gb = df.groupby("year").agg({"hotstar_id":"count"})
gb
```

```python
plt.figure(figsize = (10,4))
sns.lineplot(x = gb.index, y = gb["hotstar_id"])
plt.show()
```
